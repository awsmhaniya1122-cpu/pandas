```python
import matplotlib.pyplot as plt
# Line chart
x = ["volleyball", "cricket", "badminton"]
y = [35, 23, 15]

plt.plot(x, y, marker='o', color='maroon')
plt.title("Line chart")
plt.xlabel("Day")
plt.ylabel("Sports")
plt.show()
```

```python
import numpy as np
# Histogram

data = np.random.normal(50, 10, 200)

plt.hist(data, bins=20, color='black')
plt.title("Histogram")
plt.show()
```

```python
# Bar chart
subjects = ['Math', 'Science', 'English']
marks = [80, 90, 70]

plt.bar(subjects, marks, color=['blue', 'green', 'orange'])
plt.title("Marks by Subject")
plt.ylabel("Marks")
plt.show()
```

```python
# Pie chart
labels = ['Mobile', 'Desktop', 'Tablet']
values = [50, 35, 15]

plt.pie(values, labels=labels, autopct='%1.0f%%')
plt.title("Device Usage")
plt.show()

print("✅ Day 1 done!")
```
